```python
from matplotlib import pyplot as plt
import pandas as pd
from IPython.display import HTML
from money_squirrel.bin import utils

%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 7)

COINS = utils.load_config("coins")

def display_all(df):
    return HTML(df.to_html())
```

```python
gdax_price = utils.query_pg("select * from gdax_gdaxprice")
interestovertime = utils.query_pg("select * from trends_interestovertime")
interestbyregion = utils.query_pg("select * from trends_interestbyregion")
```

```python
gdax_price.head()
```

```python
interestovertime.sample(10).T
```

```python
try:
    for c in COINS:
        interestovertime[c] = interestovertime['scores'].apply(lambda d: d.get(COINS[c]))
        interestbyregion[c] = interestbyregion['scores'].apply(lambda d: d.get(COINS[c]))
except KeyError, ValueError:
    pass
        
try:
    interestbyregion.drop('scores', axis=1, inplace=True)
    interestovertime.drop('scores', axis=1, inplace=True)
except KeyError, ValueError:
    pass

try:
    gdax_price.set_index('id', inplace=True)
    interestbyregion.set_index('id', inplace=True)
    interestovertime.set_index('id', inplace=True)
except KeyError, ValueError:
    pass
```

```python
interestbyregion.head()
```

```python
interestbyregion.groupby('geo').describe().unstack()['BTC'].sort_values('mean', ascending=False)
```

```python
interestovertime.groupby('end_dt')[COINS.keys()].mean().plot()
```

```python
interestovertime.set_index('dt')[COINS.keys()].sort_index().rolling(1000).mean().plot();
```

```python
gdax_price.drop_duplicates(['product', 'dt', 'volume'], inplace=True)
```

```python
gdax_price.drop_duplicates(['product', 'dt', 'volume'])\
    .pivot(index='dt', columns='product', values='volume')\
    .rolling(6).mean()\
    .plot()
```

```python
gdax_price.drop_duplicates(['product', 'dt', 'close'])\
    .pivot(index='dt', columns='product', values='close')\
    .rolling(6).mean()\
    .plot()
```

```python
gdax_price.head()
```

```python
df = interestovertime.assign(day=lambda df: df.dt.apply(lambda z: z.date()))\
    .groupby('day')['BTC'].mean().to_frame()\
    .merge(
        gdax_price[gdax_price['product']=='BTC-USD']\
            .assign(day=lambda df: df.dt.apply(lambda z: z.date()))\
            .groupby('day')[['close', 'volume']].mean(),
        left_index=True, right_index=True
    ).assign(next_close = lambda df: df['close'].shift(1))\
    .assign(next_volume = lambda df: df['volume'].shift(1))\
    .assign(close_delta = lambda df: df['next_close'] - df['close'])
    
df.head()
```

```python
df.plot(kind='scatter', x='BTC', y='close_delta')
```
